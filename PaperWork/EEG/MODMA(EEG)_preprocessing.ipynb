```python
################ extract raw matfile from folder and convert into numpy array
import os
import pandas as pd
import scipy.io as sio
import re
import numpy as np
################################ MDD class dataset convert raw mat file into numpy array ###################
datapath="EEG_128channels_resting_lanzhou_2015"  ### use yoour path
save_path='MDD' ################## save MMD data
lstpath=os.listdir(datapath)
patren='0201' ################ slect this pattern blong to MMD class as shown in xcel file
mdd_list=list(filter(lambda x: patren in x,lstpath))
for mdd in mdd_list:
    filepath=os.path.join(datapath,mdd)  #read path
    #print(filepath)
    datafile=sio.loadmat(filepath) #### read matfile
    r = re.compile(".*a0.*") ##### slect eeg array, we have different pattern inside the mat file
    dataff=list(filter(r.match, datafile))
    dataff1=dataff[0]
    #print(dataff1.shape)
    dataeeg=datafile[dataff1]  ###### slect eeg data array
    np.save(os.path.join(save_path,dataff1[1:9]+".npy"),dataeeg)  #### save eeg numpy array into drive
```
