**<h2>Anaconda prompt 접속 (혹시 몰라서 관리자 권한으로 접속함)</h2>**

**<h3>Anaconda prompt에서 커맨드 작성</h3>**

**1. conda create --name tensorflowgpu python = 3.10**

**2. conda activate tensorflowgpu**

**3. conda install -c conda-forge cudatoolkit=11.2 cudnn=8.1.0**

**4. pip install tensorflow==2.10**



**1,2 에서 tensorflowgpu는 가상환경 이름이므로 편한대로 설정하면됨**

**3 에서 cudatoolkit이랑 cudnn 버전은 위에서 확인해뒀던 버전으로 숫자 변경**

**4 에서 역시 확인해뒀던 tensorflow 버전으로 숫자 변경**

**pip install jupyter numpy pandas matplotlib seaborn scikit-learn tensorflow keras statsmodels scipy nltk gensim spacy plotly bokeh folium**
