#**09-1 순차 데이터와 순환 신경망**

# 순차 데이터
> 순차 데이터 (sequential data)는 텍스트나 시계열 데이터 (time series data)와 같이 순서에 의미가 있는 데이터를 말함
- 순차 데이터를 다룰 때는 이전에 입력한 데이터를 기억하는 기능이 필요함

> 피드 포워드 신경망  (feedforward nearal network, FFNN): 입력 데이터의 흐름이 앞으로만 전달되는 신경망
- 하나의 샘플을 사용하여 정방향 계산을 수행하고 나면 그 샘플은 버려지고 다음 샘플을 처리할 때 재사용 X
- 완전 연결 신경망, 합성곱 신경망 등

# 순환 신경망
> 순환 신경망 (recurrent neural network, RNN): 다음 샘플을 위해서 이전 데이터가 신경망 층에 순환되는 신경망
- 완전 연결 신경망에서 완전 연결 신경망 이전 데이터의 처리 흐름을 순환하는 고리하나를 추가

<img src= "https://drive.google.com/uc?export=view&id=1Icu4Mx5uuQ1f6PevzGOFeXclqsw9wCdH" width=700 height=500>

- 뉴런의 출력이 다시 자기 자신으로 전달됨 (어떤 샘플 처리시 이전에 사용했던 데이터 재사용)

> 예를 들어, A,B,C, 3개의 샘플을 처리하는 순환 신경망의 뉴런이 있다고 가정하고 O는 출력된 결과이다.

1. 샘플 A를 처리하고 난 출력(Oa)이 다시 뉴런으로 돌아옴
 - Oa에는 A에 대한 정보가 포함됨
2. 샘플 B를 처리할 때, Oa를 함께 사용
 - Oa와 B를 사용해서 만든 Ob에는 A에 대한 정보도 어느정도 포함됨
3. C를 처리할 때에는 Ob를 사용
 - Oc에는 Ob를 사용했으므로, B에 대한 정보와 A에 대한 정보가 모두 포함됨
 - A에 대한 정보보다는 B에 대한 정보가 더 많음

> 타임스텝 (timestep): 앞에서 처럼 샘플을 처리하는 한 단계를 말함
- 순환 신경망은 이전의 타임스텝의 샘플을 기억하지만, 타임 스텝이 오래될 수록 순환되는 정보는 희미해짐

> 셀 (cell): 순환 신경망에서 특별하게 층을 부르는 말
- 한 셀에는 여러 개의 뉴런이 있지만 완전 연결 신경망과 달리 뉴런을 모두 표시하지 않고 하나의 셀로 층을 표현함

> 셀의 출력을 은닉상태 (hidden state)라고 부름
- 신경망의 구조마다 부르는 이름이 다를 수 있지만 기본 구조는 같음
 - 입력에 어떤 가중치를 곱하고 활성화 함수를  통과시켜 다음 층으로 보냄
 - 다른 점은 층의 출력(즉, 은닉상태)을 다음 타임 스텝에 재사용하는 것

> 일반적으로 은닉층의 활성화 함수로는 하이퍼블릭 탄젠트 (hyperbolic tangent) 함수인 tanh가 많이 사용됨
- tanh 함수도 S자 모양을 띠기 때문에 종종 시그모이드 함수라고 부르기도 함
- tanh 함수는 시그모이드와는 달리 -1~1 사이의 범위를 가짐

> 순환 신경망에서도 피드포워드 신경망에서와 같이 뉴런은 입력과 가중치를 곱함, 다만 순환 신경망의 뉴런은 가중치가 하나 더 존재
- 바로 이전 타임스텝의 은닉 상태에 곱해지는 가중치
- 셀은 입력과 이전 타임스텝의 은닉 상태를 사용하여 현재 타임스텝의 은닉 상태를 만듬
>

> Wx는 입력에 곱해지는 가중치이고 Wh는 이전 타임스텝의 은닉 상태에 곱해지는 가중치

> 피드포인트 신경망과 마찬가지로 뉴런마다 하나의 절편이 포함됨

<img src= "https://drive.google.com/uc?export=view&id=1XlmqmK6Z7frpApre4v1FMwNDCMdjDjmS" width=700 height=400>


1. 타임스텝 1에서 셀의 출력 h1이 타임스텝 2의 셀로 주입됨
 - 이때,Wh와 곱해짐
2. 타임스텝 2에서 셀의 출력 h2가 타임스텝 3의 셀로 주입됨
 0 이때에도 wh와 곱해짐

> 여기서 알 수 있는 것은 모든 타임스텝에서 사용되는 가중치는 Wh 하나라는 점임
- 가중치 Wh는 타임스텝에 따라 변화되는 뉴런의 출력을 학습함
 - 이런 능력이 순차 데이터를 다루는데 필요

> 타임스텝 1에서 사용되는 이전 은닉 상태 h0는 간단하게 0으로 초기화함

# 셀의 가중치와 입출력

> 순환층에 입력되는 특성의 개수가 4개이고 순환층의 뉴런이 3개라고 가정
- Wx의 크기는 입력층과 순환층의 뉴런이 모두 연결되기 때문에 4 x 3 = 12개

<img src= "https://drive.google.com/uc?export=view&id=1UghS-msT-j13QFlilZyfxu_FkB4syEQZ" width=700 height=400>

> 다음 타임스텝에 재사용되는 은닉 상태를 위한 가중치 Wh의 크기
- 순환층에 있는 첫 번째 뉴런(r1)의 은닉 상태가 다음 타임스텝에 재사용될 때 첫 번째 뉴런과 두 번째 뉴런, 세 번째 뉴런에 모두 전달 됨 (빨간색 선)
- 두 번재 뉴런의 은닉 상태 역시 모든 뉴런에 모두 전달 되고 (파란색 선)
- 세 번째 뉴런의 은닉 상태 역시 동일함 (검은색 선)
- 따라서 이 순환층에서 은닉 상태를 위한 가중치 Wh는 3x3 = 9개


> 모델 파라미터 개수 계산 = 가중치 + 절편  = Wx + wh + 절편
- 각 뉴런마다 하나의 절편이 존재하므로 이 순환층은 모두 12 + 9 +3 = 24개의 모델 파라미터가 존재

> 순환층은 일반적으로 샘플마다 2개의 차원을 가짐
- 보통 하나의 시퀸스 (sequence)라고 말함
- 시퀸스 안에는 여러 개의 아이템이 들어 있음
  - 여기에서 시퀸스의 길이가 타임스텝의 길이가 됨

> 예를 들어 "I am a boy"란 문장이 들어 있다고 가정
 - 이 샘플은 4개의 단어로 이루어짐
 - 각 단어를 3개의 어떤 숫자로 표현한다고 가정 (1,4,3)

> 이런 입력이 순환층을 통과하면 두 번째, 세 번째 차원이 사라지고 순환층의 뉴런 개수만큼 출력 됨  

<img src= "https://drive.google.com/uc?export=view&id=1g2pKNxEr8GiywYteC6_6QsyppqD8iSZk" width=700 height=400>


> 하나의 샘플은 시퀸스 길이(여기서는 단어 개수)와 단어 표현의 2차원 배열
- 앞에서는 셀이 모든 타임스텝에서 출력을 만든 것처럼 표현함
- 하지만 사실 순환층은 기본적으로 마지막 타임스텝의 은닉 상태만 출력으로 내보냄

> 다음 그림에서 이런 특징을 그림으로 표현하기 위해 셀의 출력을 점섬으로 표시하고 마지막 타임스텝의 은닉 상태임을 나타내기 위해 아랫첨자 f를 사용함

<img src= "https://drive.google.com/uc?export=view&id=1fGE4clkNAjt_jQN3AUVsQeiY7K2n-HpS" width=700 height=400>

> 이는 마치 입력된 시퀸스 길이를 모두 읽어서 정보를 마지막 은닉 상태에 압축하여 전달하는 것처럼 볼 수 있음

> 순환 신경망도 완전 연결 신경망이나 합성곱 신경망 처럼 여러 개의 층을 쌓을 수 있음
- 순환층을 여러 개 쌓았을 때, 셀의 입력은 샘플마다 타임스텝과 단어 표현으로 이루어진 2차원 배열이여야 함
 - 따라서 첫 번째 셀이 마지막 타임스텝의 은닉 상태만 출력해서는 안됨
 - 이런 경우, 마지막 셀을 제외한 다른 모든 셀은 모든 타임스텝의 은닉 상태를 출력함
 - 첫 번째 셀은 모든 타임스텝의 은닉 상태를 출력하고, 두 번째 셀은 마지막 타임스텝의 은닉 상태만 출력함

> 합성곱 신경망과 마찬가지로 순환 신경망도 마지막에는 밀집층을 두어 클래스를 분류함
- 다중 분륭일 경우는 출력층에 클래스 개수만큼 뉴런을 두고 소프트맥스 활성화 함수 사용
- 이진 분류일 경우에는 하나의 뉴런을 두고 시그모이드 활성화 함수 사용
- 합성곱 신경망과 다른점은 마지막 셀의 출력이 1차원이기 때문에 Flatten 클래스로 펼칠 필요 X

>ex) 다중 분류 문제에서 입력 샘플의 크기가 (20, 100)일 경우 하나의 순환층을 사용하는 순환 신경망의 구조

> 이 예에서 샘플은 20개의 타임 스텝으로 이루어져 있고 각 타임스텝은 100개의 표현 또는 특성으로 이루어져 있음

> 이 샘플이 순환층의 셀을 통과하면 모든 타임스텝을 처리하고 난 후의 은닉상태만 출력됨
- 이 은닉 상태의 크기는 셀에 있는 뉴런의 개수가 되므로 (10,)임



> 샘플마다 셀이 1차원 배열을 출력하기 때문에 합성곱 신경망 처럼 Flatten 클래스로 펼칠 필요없이 바로 출력층에 연결할 수 있음
