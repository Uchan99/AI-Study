```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
df_red = pd.read_csv("C:/AI_study/winequality-red.csv", sep=';')
df_red
```

```python
df_white = pd.read_csv("C:/AI_study/winequality-white.csv",sep=';')
df_white
```

```python
df = pd.concat([df_red, df_white], axis=0)
df
```

```python
y = df['residual sugar']
y
```

```python
x= df.drop('residual sugar',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=423)
```

```python
from sklearn.svm import SVR

clf_svm = SVR()
clf_svm.fit(X_train, y_train)

pred_svm = clf_svm.predict(X_test)

print("\n--- SVM Regressor ---")
print(pred_svm)
print(clf_svm.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_svm, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.linear_model import LinearRegression

clf_lr = LinearRegression()
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Regressor ---")
print(pred_lr)
print(clf_lr.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_lr, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeRegressor

clf_dt = DecisionTreeRegressor()
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Regressor ---")
print(pred_dt)
print(clf_dt.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_dt, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_clf = RandomForestRegressor()
rf_clf.fit(X_train, y_train)

pred_rf = rf_clf.predict(X_test)

print ("\n--- Random Forest Regressor ---")
print(pred_rf)
print(rf_clf.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_rf, y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.neighbors import KNeighborsRegressor

# KNN 회귀 모델 생성
clf_knn = KNeighborsRegressor()
clf_knn.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
pred_knn = clf_knn.predict(X_test)

# 훈련 데이터에 대한 모델의 성능 출력 (R-squared)
print ("\n--- KNeighbors Classifier ---")
print(pred_knn)
print(clf_knn.score(X_train, y_train))
```

```python
mse = np.sqrt(mean_squared_error(pred_knn, y_test))
print('평균제곱근오차', mse)
```
